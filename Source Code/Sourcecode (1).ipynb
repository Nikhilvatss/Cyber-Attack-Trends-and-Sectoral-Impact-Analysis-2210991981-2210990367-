```python
df["Actor Type"] = df["Actor Type"].fillna("Unknown")

df["Estimated Damage"] = df["Estimated Damage"].astype(str)
```

```python
actor_counts = df["Actor Type"].value_counts()

actor_counts.plot(kind='bar')
plt.title("Number of Attacks by Actor Type")
plt.xlabel("Actor Type")
plt.ylabel("Count")
plt.show()
```

```python
attack_counts = df["Attack Type"].value_counts()

attack_counts.plot(kind='bar')
plt.title("Distribution of Attack Types")
plt.xlabel("Attack Type")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
sector_counts = df["Target Sector"].value_counts()

sector_counts.plot(kind='bar')
plt.title("Targeted Sectors in Cyber Attacks")
plt.xlabel("Sector")
plt.ylabel("Number of Attacks")
plt.xticks(rotation=45)
plt.show()
```

```python
impact_counts = df["Impact Level"].value_counts()

impact_counts.plot(kind='bar')
plt.title("Impact Level of Cyber Attacks")
plt.xlabel("Impact Level")
plt.ylabel("Count")
plt.show()
```

```python
year_counts = df["Year"].value_counts().sort_index()

year_counts.plot(kind='line', marker='o')
plt.title("Cyber Attacks Over Time")
plt.xlabel("Year")
plt.ylabel("Number of Attacks")
plt.show()
```

```python
country_counts = df["Target Country"].value_counts().head(10)

country_counts.plot(kind='bar')
plt.title("Top Target Countries")
plt.xlabel("Country")
plt.ylabel("Number of Attacks")
plt.xticks(rotation=45)
plt.show()
```

```python
pd.crosstab(df["Actor Type"], df["Impact Level"]).plot(kind='bar')

plt.title("Actor Type vs Impact Level")
plt.xlabel("Actor Type")
plt.ylabel("Count")
plt.show()
```

```python
pd.crosstab(df["Attack Type"], df["Impact Level"]).plot(kind='bar')

plt.title("Attack Type vs Impact Level")
plt.xlabel("Attack Type")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
pd.crosstab(df["Target Sector"], df["Impact Level"]).plot(kind='bar')

plt.title("Sector vs Impact Level")
plt.xlabel("Sector")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
high_impact = df[df["Impact Level"].isin(["High", "Critical"])]

high_counts = high_impact["Actor Type"].value_counts()

high_counts.plot(kind='bar')
plt.title("High & Critical Attacks by Actor Type")
plt.xlabel("Actor Type")
plt.ylabel("Count")
plt.show()
```
